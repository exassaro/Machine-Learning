```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer



df=pd.read_csv('C:/Users/USER/Documents/Github/Machine Learning/datasets/spam_mail_classifier.csv')
```

```python
df['label_enc']=LabelEncoder().fit_transform(df['label'])
```

```python
X_train,X_test,y_train,y_test=train_test_split(df['email_text'],df['label_enc'],test_size=0.2,random_state=42)
```

```python
df.head()
```

```python
vectorizer=TfidfVectorizer()
X_train_vec=vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
rand=RandomForestClassifier(max_depth=6,n_estimators=100,random_state=42)
rand.fit(X_train_vec,y_train)

```

```python
y_pred=rand.predict(X_test_vec)

print(classification_report(y_pred,y_test))
```
