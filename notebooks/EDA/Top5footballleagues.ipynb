### Multivariate Analysis

```python
#Top Performers

top5=df.sort_values("G+A",ascending=False).head(5)[["Player", "Gls", "Ast", "xG", "xAG"]]
print(top5)

#This is optional spider graph/Radar plot for each player
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Select top 5 players and relevant stats
stats = ["Gls", "Ast", "xG", "xAG"]
top5 = df.sort_values("G+A", ascending=False).head(5)[["Player"] + stats]

# Step 2: Normalize the stats (min-max scaling)
data = top5.copy()
data[stats] = (data[stats] - data[stats].min()) / (data[stats].max() - data[stats].min())

# Step 3: Radar chart setup
labels = stats
num_vars = len(labels)

# Compute angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # close the loop

# Setup the plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot each player's line
for i, row in data.iterrows():
    values = row[stats].tolist()
    values += values[:1]  # close the loop
    ax.plot(angles, values, label=row["Player"])
    ax.fill(angles, values, alpha=0.1)

# Style
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), labels)
ax.set_title("Top 5 Players Radar (Normalized Stats)", y=1.1)
ax.set_ylim(0, 1)
ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))

plt.tight_layout()
plt.show()
```
