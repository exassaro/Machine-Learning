# Pandas Series

```python
import pandas as pd

a=[1,2,3,4]
myvar=pd.Series(a)
print(myvar) # gives kind of index num if nothing specified and type of variable
```

```python
a=[1,7,90]
myvar=pd.Series(a,index=["x","y","z"])
print(myvar)
print(myvar["x"])#creating custom labels
```

```python
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)#creating series with key value pairs
```

```python

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ["day1", "day2"])

print(myvar) #creating series only on specific data as above mentioned
```

# Data Frames

```python
import pandas as pd

data={
    "calories":[421,344,756],
    "durataion":[40,20,90]
}
df=pd.DataFrame(data)
print(df)
#refer to the row index to locate:
print(df.loc[0])#this return series
print(df.loc[[0,1]])#this return dataframe
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"]) #named index

print(df) 
#refer to the named index:
print(df.loc["day2"])
```

# Cleaning empty cells

```python
import pandas as pd

df = pd.read_csv('C:/Users/USER/Documents/Github/Machine Learning/datasets/data2.csv')

df.fillna(130, inplace = True) #Notice in the result: empty cells got the value 130 (in row 18, 22 and 28).
print(df.to_string())
```

```python
import pandas as pd

df = pd.read_csv('C:/Users/USER/Documents/Github/Machine Learning/datasets/data2.csv')

x=df.fillna({"Calories": 130})

print(x.to_string())

#This operation inserts 130 in empty cells in the "Calories" column (row 18 and 28).
```

```python
import pandas as pd
df = pd.read_csv('C:/Users/USER/Documents/Github/Machine Learning/datasets/data2.csv')
x = df["Calories"].mean()
df.fillna({"Calories": x}, inplace=True)
print(df.to_string())
print(df.to_string) #As you can see in row 18 and 28, the empty values from "Calories" was replaced with the mean: 304.68
```

```python
import pandas as pd
df = pd.read_csv('C:/Users/USER/Documents/Github/Machine Learning/datasets/data2.csv')
x = df["Calories"].median()
df.fillna({"Calories": x}, inplace=True)
print(df.to_string())
#As you can see in row 18 and 28, the empty values from "Calories" was replaced with the median: 291.2
```

```python
import pandas as pd
df = pd.read_csv('C:/Users/USER/Documents/Github/Machine Learning/datasets/data2.csv')
x = df["Calories"].mode()[0]
df.fillna({"Calories": x}, inplace=True)
print(df.to_string())
#As you can see in row 18 and 28, the empty value from "Calories" was replaced with the mode: 300.0
```

# Clean Wrong data

```python
import pandas as pd

df = pd.read_csv('C:/Users/USER/Documents/Github/Machine Learning/datasets/data4.csv')

print(df.to_string())

df.loc[7,'Duration'] = 45 #it is most likely a typo, and the value should be "45" instead of "450"

print(df.to_string())
```

```python
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120

print(df.to_string()) #Loop through all values in the "Duration" column
```

```python
df = pd.read_csv('C:/Users/USER/Documents/Github/Machine Learning/datasets/data.csv')

for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.drop(x, inplace = True)

#Delete rows where "Duration" is higher than 120

#remember to include the 'inplace = True' argument to make the changes in the original 
# DataFrame object instead of returning a copy
```
